```python
import tkinter as tk
from tkinter import *
from tkinter import filedialog
from tkinter.ttk import Combobox
import pyttsx3
import os
from gtts import gTTS
from playsound import playsound
import PyPDF2



root = Tk()
root.title("Text to speech converter")
root.geometry("1000x580")
root.resizable(False,False)
root.configure(bg="#ffcce0")


tts = pyttsx3.init()
# upload pdf file



def UploadAction(event=None):
    filename = filedialog.askopenfilename()
    
    openfile=open(filename,'rb')
    pdf_Reader = PyPDF2.PdfReader(openfile)
    num_page = len(pdf_Reader.pages)
    pdftext=""
    for i in range(num_page):
        page = pdf_Reader.pages[i]
        pdftext = page.extract_text()
     
    
        text_box.insert(END, pdftext)
  
    

def speaknow():
   
    text = text_box.get(1.0, END)
    gender = gender_box.get()
    speed = speed_box.get()
    sound = sound_box.get()
    voices = tts.getProperty('voices')
    Volume = tts.getProperty('volume')

#     tts.save_to_file(text,'audio.mp3')
    


  
    
    def setvoice():
        if(gender == 'Male'):
            tts.setProperty('voice',voices[0].id)
            
            tts.say(text)
            tts.runAndWait()
            tts.stop()

        else:
            tts.setProperty('voice',voices[1].id)
            tts.say(text)
            tts.runAndWait()
            tts.stop()

    
    if(text):
        if(speed == 'Fast'):
            tts.setProperty('rate',200)
        elif(speed == 'Medium'):
            tts.setProperty('rate',125)
        else:
            tts.setProperty('rate',50)
            
    if(text):
        if(sound == 'High'):
            tts.setProperty('volume',1.0) 
        elif(sound == 'Medium'):
            tts.setProperty('volume',0.5)
        else:
            tts.setProperty('volume',0.2)
    setvoice()


    
 

def Exit():
    root.destroy()
    
def fileSave( lang='en', filename=input("enter a file name")+".mp3"):
    text = text_box.get(1.0, END)
   
    tts = gTTS(text=text, lang=lang)
    tts.save(filename)

 
                       
text_box = Text(root,font="calibri 20", bg="white",relief=GROOVE, wrap=WORD,bd=0)
text_box.place(x=30,y=150,width=940,height=180)
# Combobox for gender
gender_box = Combobox(root, values=['Male', 'Female'], state='readonly', font='arial 14 bold', width=23)
gender_box.set('Male')
gender_box.place(x=30, y=380)

# Combobox for speed
speed_box = Combobox(root, values=['Fast', 'Medium', 'Slow'], state='readonly', font='arial 14 bold', width=23)
speed_box.set('Medium')
speed_box.place(x=350, y=380)

# Combobox for sound
sound_box = Combobox(root, values=['High', 'Medium', 'Low'], state='readonly', font='arial 14 bold', width=23)
sound_box.set('Medium')
sound_box.place(x=670, y=380)

Label(root,text='Select voice', font='TimesNewRoman 15 bold', bg='#C70039', fg='White').place(x=30,y=350)
Label(root,text='Select speed', font='TimesNewRoman 15 bold', bg='#C70039', fg='White').place(x=350,y=350)
Label(root,text='Select Volume', font='TimesNewRoman 15 bold', bg='#C70039', fg='White').place(x=670,y=350)



upload_btn = Button(root, text="File upload", compound=LEFT, bg='white', width=25, font='arial 14 bold',borderwidth='0.1c', command=UploadAction)
upload_btn.place(x=30, y=100)
save_btn = Button(root, text="save", compound=LEFT, bg='white', width=25, font='arial 14 bold',borderwidth='0.1c', command=fileSave)
save_btn.place(x=300, y=100)




play_btn=Button(root,text="Play",compound=LEFT,bg='#C70039',fg='White',width=20,font="arial 14 bold",borderwidth='0.1c',command= speaknow)
play_btn.place(x=380,y=450)

stop_btn=Button(root,text="stop",compound=LEFT,bg='#C70039',fg='White',width=20,font="arial 14 bold",borderwidth='0.1c',command=Exit)
stop_btn.place(x=180,y=450)





root.mainloop()


```
